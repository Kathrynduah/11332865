```python
import matplotlib.pyplot as plt
plt.hist(Dataset['quality'], bins=10)  
plt.title('Quality Distribution')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(7,6))
plt.hist(Dataset['quality'],bins=6,alpha=0.6, color='black')
plt.title('Dristribution of Wine Quantity')
plt.xlabel('Quality')
plt.ylabel('Frquency')
plt.show()
```

```python
plt.boxplot(Dataset['residual sugar'])
plt.title('residual sugar Distribution')
plt.xlabel('residual sugar')
plt.show()
```

```python
plt.scatter(Dataset['alcohol'],Dataset['pH'])
plt.title('Alcohol vs pH')
plt.xlabel('Alcohol')
plt.ylabel('pH')
plt.show()
```
